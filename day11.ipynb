```python
x=[56,1,23,78]
y=[45,99,2,4]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, '--')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, '*')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, '3')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, 'x')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, '>')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, 'y')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, '*:y')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, '*:y', ms=200)
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
plt.plot(x,y, '*:y', ms=10, mec='r', mfc='g')
plt.ylabel("richi")
plt.xlabel("ii")
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
z=[10,11,12,13]
```

```python
plt.plot(x,y, '*:b', ms=10, mec='r', mfc='g')
plt.plot(x,z, 'o:g', ms=10, mec='g', mfc='r')

plt.show()
```

```python
plt.plot(x,y, '*:b', ms=10, mec='r', mfc='g')
plt.plot(x,z, 'o:g', ms=10, mec='g', mfc='r')
plt.ylabel("y axis",fontdict={'color':'blue','size':20})
plt.xlabel("x axis")
plt.show()
```

```python
plt.plot(x,y, '*:b', ms=10, mec='r', mfc='g')
plt.plot(x,z, 'o:g', ms=10, mec='g', mfc='r')
plt.ylabel("y axis",fontdict={'color':'blue','size':20})
plt.xlabel("x axis",fontdict={'color':'GREEN','size':20})
plt.show()
```

```python
plt.plot(x,y, '*:b', ms=10, mec='r', mfc='g')
plt.plot(x,z, 'o:g', ms=10, mec='g', mfc='r')
plt.ylabel("y axis",fontdict={'family':'serif','color':'blue','size':20})
plt.xlabel("x axis",fontdict={'color':'GREEN','size':20})
plt.title("graghyy")
plt.show()
```
